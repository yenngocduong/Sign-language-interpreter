```python
video_full_name_list = os.listdir('/content/drive/My Drive/EE8101_Deep_Learning/NaN_to_Zero_Arrays/') # Import numpy array
video_full_name_list.sort()
print(len(video_full_name_list)) # Should be the same as no. of .mp4 videos = 17875
print(video_full_name_list)
```

```python
video_trimmed_name_list = []

for name in video_full_name_list:
  video_trimmed_name_list.append(name[:-4].split('_')[1])  # Delete the number in front of each file name and the ".npy" at the end of each file name

video_trimmed_name_list.sort()
print(len(video_trimmed_name_list)) # Should be the same as no. of .mp4 videos = 17875
print(video_trimmed_name_list)
```

```python
length = len(video_trimmed_name_list)
print(length)
```

```python
words_with_at_least_16_videos = []
i = 0
while i < length:
  j = 1
  while j < 100 and (i+j) < length:
    if video_trimmed_name_list[i+j] == video_trimmed_name_list[i]:
      j += 1
    else:
      break
  if j >= 1:
    words_with_at_least_16_videos.append(video_trimmed_name_list[i])
  i = i+j

print(len(words_with_at_least_16_videos))
print(words_with_at_least_16_videos)
```

```python
words_with_at_least_16_videos = []
i = 0
while i < length:
  j = 1
  while j < 100 and (i+j) < length:
    if video_trimmed_name_list[i+j] == video_trimmed_name_list[i]:
      j += 1
    else:
      break
  if j >= 16:
    words_with_at_least_16_videos.append(video_trimmed_name_list[i])
  i = i+j

print(len(words_with_at_least_16_videos))
print(words_with_at_least_16_videos)
```

```python
video_list = []
for word in words_with_at_least_16_videos:
  for name in video_full_name_list:
    if name[:-4].split('_')[1] == word:
      video_list.append(name)
print(len(video_list))
print(video_list)
```
